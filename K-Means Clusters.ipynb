```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv("Dataset/Iris.csv")
```

```python
iris.drop(['Id'], axis=1, inplace=True)
```

```python
iris.head(3)
```

```python
SepalLengthCm = iris.iloc[:, 0].values
SepalWidthCm = iris.iloc[:, 1].values
Species = iris.iloc[:, 4].values
```

```python
np.unique(Species)
```

```python
color = {'Iris-setosa':'#00FF00', 'Iris-versicolor':'#0000FF', 'Iris-virginica':'#FF0000' }
```

```python
colors = []
for sp in Species:
    colors.append(color[sp])
del sp
```

```python
colors = np.array(colors)
```

```python
plt.scatter(SepalWidthCm, SepalLengthCm, c= colors)
plt.xlabel("Sepal Width in Cm")
plt.ylabel("Sepal Length in Cm")
plt.show()
```

```python
PetalLengthCm = iris.iloc[:, 2].values
PetalWidthCm = iris.iloc[:, 3].values
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c= colors)
plt.xlabel("Petal Width in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
```

```python
Petal = []
for i in range(len(Species)):
    Petal.append([PetalWidthCm[i], PetalLengthCm[i]])
del(i)
```

```python
Petal = np.array(Petal)
```

```python
kMeans = KMeans(n_clusters = 3)
```

```python
kMeans.fit(Petal)
predict = kMeans.predict(Petal)
```

```python
color2 = ['#00FF00', '#FF0000', '#0000FF']
```

```python
colors = []
for p in predict:
    colors.append(color2[p])
del p
```

```python
kMeans.cluster_centers_
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c= colors)
plt.scatter(kMeans.cluster_centers_[:,0], kMeans.cluster_centers_[:,1], c= '#000000', marker= 'x')
plt.xlabel("Petal Width in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
```

```python
Sepal = []
for i in range(len(Species)):
    Sepal.append([SepalWidthCm[i], SepalLengthCm[i]])
del(i)
```

```python
Sepal = np.array(Sepal)
```

```python
KMeas2 = KMeans(n_clusters=3)
predict2 = KMeas2.fit_predict(Sepal)
```

```python
colors = []
for p in predict2:
    colors.append(color2[p])
del p
```

```python
plt.scatter(SepalWidthCm, SepalLengthCm, c= colors)
plt.scatter(KMeas2.cluster_centers_[:,0], KMeas2.cluster_centers_[:,1], c= '000000', marker= 'x')
plt.xlabel("Sepal Width in Cm")
plt.ylabel("Sepal Length in Cm")
plt.show()
```
