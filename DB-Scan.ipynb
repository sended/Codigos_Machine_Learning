```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
iris = pd.read_csv("Dataset/Iris.csv")
iris.drop(['Id'], axis=1, inplace=True)
iris.head(3)
```

```python
scale_iris = iris.iloc[:, :4].values
```

```python
SSC = StandardScaler()
```

```python
scale_iris = SSC.fit_transform(scale_iris)
```

```python
SepalLengthCm = scale_iris[:, 0]
SepalWidthCm = scale_iris[:, 1]
PetalLengthCm = scale_iris[:, 2]
PetalWidthCm = scale_iris[:, 3]
Species = iris.iloc[:, 4].values
```

```python
np.unique(Species)
color = {'Iris-setosa':'#00FF00', 'Iris-versicolor':'#0000FF', 'Iris-virginica':'#FF0000' }
colors = []
for sp in Species:
    colors.append(color[sp])
del sp
```

```python
plt.scatter(SepalWidthCm, SepalLengthCm, c= colors)
plt.xlabel("Sepal Width in Cm")
plt.ylabel("Sepal Length in Cm")
plt.show()
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c= colors)
plt.xlabel("Petal Width in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
```

```python
Petal = []
for i in range(len(PetalWidthCm)):
    Petal.append([PetalWidthCm[i], PetalLengthCm[i]])
Petal = np.array(Petal)
del(i)
```

```python
dbs = DBSCAN(eps= 0.4, min_samples= 3)
```

```python
predict = dbs.fit_predict(Petal)
```

```python
predict
```

```python
color = ['#FF0000', '#0000FF']
colors = []
for p in predict:
    colors.append(color[p])
del(p)
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c=colors)
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalLengthCm")
plt.show()
```
