```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
iris = pd.read_csv("Dataset/Iris.csv")
iris.drop(['Id'], axis=1, inplace=True)
iris.head(3)
```

```python
SepalLengthCm = iris.iloc[:, 0].values
SepalWidthCm = iris.iloc[:, 1].values
PetalLengthCm = iris.iloc[:, 2].values
PetalWidthCm = iris.iloc[:, 3].values
Species = iris.iloc[:, 4].values
```

```python
np.unique(Species)
```

```python
color = {'Iris-setosa':'#00FF00', 'Iris-versicolor':'#0000FF', 'Iris-virginica':'#FF0000' }
```

```python
colors = []
for sp in Species:
    colors.append(color[sp])
del sp
```

```python
colors = np.array(colors)
```

```python
plt.scatter(SepalWidthCm, SepalLengthCm, c= colors)
plt.xlabel("Sepal Width in Cm")
plt.ylabel("Sepal Length in Cm")
plt.show()
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c= colors)
plt.xlabel("Petal Width in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
```

```python
Petal = []
for i in range(len(Species)):
    Petal.append([PetalWidthCm[i], PetalLengthCm[i]])
del(i)
Petal = np.array(Petal)
```

```python
dendrograma = dendrogram(linkage(Petal, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity= 'euclidean', linkage= 'average')
predict = hc.fit_predict(Petal)
```

```python
predict
```

```python
colors = []
for i in range(len(Species)):
    colors.append(color2[predict[i]])
del(i)
colors = np.array(colors)
```

```python
plt.scatter(PetalWidthCm, PetalLengthCm, c= colors)
plt.xlabel("Petal Width in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
```
